```python
import pandas as pd
import datetime as dt
```

```python
fecha = dt.datetime.today().strftime('%Y-%m-%d') 
```

```python
region_gba = pd.read_csv(f'Dataframe/region_gba_{fecha}.csv')
region_pampeana = pd.read_csv(f'Dataframe/region_pampeana_{fecha}.csv')
region_noroeste = pd.read_csv(f'Dataframe/region_noroeste_{fecha}.csv')
region_noreste = pd.read_csv(f'Dataframe/region_noreste_{fecha}.csv')
region_cuyo = pd.read_csv(f'Dataframe/region_cuyo_{fecha}.csv')
region_patagonica = pd.read_csv(f'Dataframe/region_patagonica_{fecha}.csv')
```
